## Label Encoding of Class Column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()

df['Class'] = label_encoder.fit_transform(df['Class'])
```

## After encoding

```python
df['Class'].unique()
```

## Splitting Data into Features and Target

```python
input_data_x = df.iloc[: , : -1]

output_data_y = df['Class']
```

## Features Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

input_data_x = pd.DataFrame(ss.fit_transform(input_data_x) , columns=input_data_x.columns)
```

```python
input_data_x
```

## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split 

X_train , X_test , y_train , y_test = train_test_split(input_data_x , output_data_y , test_size=0.2 , random_state=40)
```

## Training the Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model = RandomForestClassifier(
    n_estimators=200,    
    max_depth=4,          
    min_samples_split=10,  
    min_samples_leaf=5,    
    max_features='sqrt',    
    bootstrap=True,         
)
```

```python
RF_model.fit(X_train , y_train)
```

## Evaluating Model Performance

```python
train_score = RF_model.score(X_train, y_train) * 100
test_score = RF_model.score(X_test, y_test) * 100
```

```python
print(f"Train Score: {train_score:.2f}%")
print(f"Test Score: {test_score:.2f}%")
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
mse = mean_squared_error(y_test, RF_model.predict(X_test))
mae = mean_absolute_error(y_test, RF_model.predict(X_test))
```

```python
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```
